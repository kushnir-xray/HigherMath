```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

def draw_funny_fig():
    fig = figure()
    ax = Axes3D(fig)
    X = np.arange(-10, 10, 1)
    Y = np.arange(-10, 10, 1)
    X, Y = np.meshgrid(X, Y)
    Z = -X**2 + Y
    Z2 = np.cos(X/3)*10 + np.cos(Y/1)*10
    ax.plot_wireframe(X, Y, Z2, color='green')
    ax.plot_wireframe(X, Y, Z, color='black')
    
    show()
    
draw_funny_fig()
```
