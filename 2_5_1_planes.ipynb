```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

def draw_parallel_planes(A=1, B=5, C=8, D=0, z_offset=1):
    fig = figure()
    ax = Axes3D(fig)
    X = np.arange(-5, 5, 0.5)
    Y = np.arange(-5, 5, 0.5)
    X, Y = np.meshgrid(X, Y)
    Z = (A*X + B*Y + D)/C
    Z2 = (A*X + B*Y + D)/C + z_offset
    ax.plot_wireframe(X, Y, Z, color='red')
    ax.plot_wireframe(X, Y, Z2, color='blue')
    ax.scatter(0, 0, 0, 'z', 50, 'black')
    show()

draw_parallel_planes(0.6, 0.8, z_offset=2)
```
